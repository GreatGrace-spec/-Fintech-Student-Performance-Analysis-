```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_2024 = [3500, 4200, 4800, 5100, 4900, 5300]
expenses = [2000, 2300, 2500, 2600, 2700, 2900]

x = np.arange(len(months))   # gives [0, 1, 2, 3, 4, 5]
width = 0.35                 # sets bar thickness and spacing

plt.bar(x - width/2, sales_2024, width, label='Sales', color='blue')
plt.bar(x + width/2, expenses, width, label='Expenses', color='orange')

plt.xticks(x, months)        # replaces 0–5 on x-axis with Jan–Jun
plt.xlabel("Month")
plt.ylabel("Amount (₦)")
plt.title("Sales vs Expenses (2024)")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
sales = [200, 250, 300, 280, 350]
profit = [50, 70, 80, 65, 90]

plt.figure(figsize=(10, 4))

# First plot
plt.subplot(1, 2, 1)  # (rows, columns, position)
plt.bar(x, sales, color='skyblue')
plt.title("Sales")
plt.xlabel("Month")
plt.ylabel("Sales Amount")

# Second plot
plt.subplot(1, 2, 2)
plt.plot(x, profit, marker='o', color='green')
plt.title("Profit")
plt.xlabel("Month")
plt.ylabel("Profit Amount")

plt.tight_layout()
plt.show()
```

```python
plt.plot(x, profit, color='green', marker='o')
plt.title("Profit Over Time")

# Annotate the highest profit point
max_index = profit.index(max(profit))
plt.annotate('Peak Profit', 
             xy=(x[max_index], profit[max_index]), 
             xytext=(x[max_index]+0.2, profit[max_index]+10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```

```python
plt.plot(x, sales)
plt.title("Sales with Custom Axes")
plt.xlabel("Quarter")
plt.ylabel("Sales ($)")

plt.xlim(0, 6)  # set range for x-axis
plt.ylim(50, 400)  # set range for y-axis
plt.xticks([1, 2, 3, 4, 5])
plt.yticks([100, 200, 300, 400])

plt.show()
```

```python
plt.style.use('ggplot')  # try 'seaborn', 'bmh', 'classic', etc.

plt.plot(x, profit, color='purple', marker='o')
plt.title("Styled Chart Example")
plt.xlabel("Month")
plt.ylabel("Profit ($)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 350],
    'Profit': [50, 70, 80, 65, 90]
}

df = pd.DataFrame(data)

# Plot a simple line chart
df.plot(x='Month', y='Sales', kind='line', title='Monthly Sales')
plt.show()

```

```python
df.plot(x='Month', y=['Sales', 'Profit'], kind='line', marker='o')
plt.title("Sales vs Profit")
plt.ylabel("Amount ($)")
plt.grid(True)
plt.show()
```

```python
df.set_index('Month')['Profit'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Profit Distribution")
plt.ylabel('')  # remove default label
plt.show()
```

```python
df['Sales'].plot(kind='hist', bins=5, color='orange', title='Sales Distribution')
plt.xlabel("Sales Amount")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset
data = {
    'Student': ['John', 'Mary', 'Alex', 'Lily', 'Ben', 'Jane', 'Sam', 'Tina'],
    'GradeClass': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'GPA': [3.8, 3.2, 3.6, 2.9, 3.4, 3.9, 2.8, 3.1],
    'StudyTimeWeekly': [12, 9, 10, 5, 8, 13, 6, 7],
    'Absences': [2, 4, 3, 7, 5, 1, 6, 4]
}

df = pd.DataFrame(data)
print(df)
```

```python
avg_gpa = df.groupby('GradeClass')['GPA'].mean().sort_index()
print(avg_gpa)
```

```python
plt.figure(figsize=(7, 5))
plt.bar(avg_gpa.index, avg_gpa.values, color='skyblue')
plt.title('Average GPA per Grade Class')
plt.xlabel('Grade Class')
plt.ylabel('Average GPA')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

colors = plt.cm.viridis(avg_gpa.values / max(avg_gpa.values))  # gradient colors

bars = plt.bar(avg_gpa.index, avg_gpa.values, color=colors)

plt.title('Average GPA by Grade Class', fontsize=14)
plt.xlabel('Grade Class')
plt.ylabel('Average GPA')

# Annotate each bar with GPA value
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.02,
             f'{height:.2f}', ha='center', fontsize=10)

plt.show()
```

```python
pd.plotting.scatter_matrix(
    df[['StudyTimeWeekly', 'Absences', 'GPA']],
    figsize=(8, 6),
    diagonal='hist',
    color='teal',
    alpha=0.7
)

plt.suptitle('Scatter Matrix: Study Time, Absences, GPA', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df.boxplot(column='StudyTimeWeekly', by='GradeClass', grid=True, patch_artist=True)

plt.title('Study Time by Grade Class')
plt.suptitle('')  # removes default title
plt.xlabel('Grade Class')
plt.ylabel('Study Time (hours per week)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

colors = ['lightcoral', 'lightblue', 'lightgreen']
df.boxplot(
    column='StudyTimeWeekly',
    by='GradeClass',
    patch_artist=True,
    boxprops=dict(facecolor='white', color='black'),
    medianprops=dict(color='red'),
)

plt.title('Study Time Distribution by Grade Class')
plt.suptitle('')
plt.xlabel('Grade Class')
plt.ylabel('Study Time (hours)')
plt.grid(True)
plt.show()
```
